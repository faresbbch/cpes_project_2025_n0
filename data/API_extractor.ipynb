```python
import csv
import requests
import pandas as pd
```

```python
# URL of the API
API_URL = "https://db.ygoprodeck.com/api/v7/cardinfo.php"

# Data extraction from the API
response = requests.get(API_URL)

if response.status_code == 200:
    data = response.json().get("data", [])  # Get the card list

    # Open a new CSV file in writing mode
    with open("my_dataset.csv", mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)

        # Write the header
        writer.writerow(["name"])

        # Fill the file
        for card in data:
            # Write the next CSV line 
            writer.writerow([
                card["name"]
            ])


API_URL_FOR_CARD = "https://db.ygoprodeck.com/api/v7/cardinfo.php?name="

df = pd.read_csv("my_dataset.csv")
df["desc"] = ""
df["atk"] = ""
df["def"] = ""
df["attribute"] = ""
df["type"] = ""
df["frametype"] = ""
df["level"] = ""

df_extention = []

for i, name in enumerate(df['name']):
    response = requests.get(API_URL_FOR_CARD + name)

    if response.status_code == 200:
        card_data = response.json().get("data", [])[0]  # Get the data of one card
        
        # Extract each parameter
        df.at[i, "desc"] = card_data.get("desc", None)
        df.at[i, "atk"] = card_data.get("atk", None)
        df.at[i, "def"] = card_data.get("def", None)
        df.at[i, "attribute"] = card_data.get("attribute", None)
        df.at[i, "type"] = card_data.get("race", None)
        df.at[i, "frametype"] = card_data.get("frameType", None)
        df.at[i, "level"] = card_data.get("level", None)

display(df)

df.to_csv("my_dataset.csv", index=False)
```
